```python
import pandas as pd
from pycaret.regression import * 
```

```python
data = pd.read_csv('tips.csv')
data
```

```python
data.describe()
```

```python
s = setup(data, target = 'tip', session_id = 123)
```

```python
best = compare_models()
```

```python
new_data = pd.DataFrame({'total_bill': [100],
                         'sex': ['Female'],
                         'smoker': ['Yes'],
                         'day': ['Sun'],
                         'time': ['Dinner'],
                         'size': [4]
                        })
new_data
```

```python
new_pred = predict_model(best, data=new_data)
new_pred
```

```python
new_pred.at[0, 'prediction_label']
```

```python
new_data2 = pd.DataFrame({'total_bill': [50],
                         'sex': ['Male'],
                         'smoker': ['Yes'],
                         'day': ['Mon'],
                         'time': ['Lunch'],
                         'size': [2]
                        })
new_data2
```

```python
new_pred2 = predict_model(best, data=new_data2)
new_pred2
```
